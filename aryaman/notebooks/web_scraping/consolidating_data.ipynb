```python
import pandas as pd
import ast
```

```python
products = pd.read_csv('aryaman/csv/categories_and_products/products.csv')
```

```python
scraped_df = pd.read_csv('aryaman/csv/scraping/scraping_details.csv',encoding='utf-8')
```

```python
concatenated_df = pd.read_csv('concatenated.csv',encoding='utf-8')
```

```python
concatenated_df
```

```python
scraped_df.isnull().sum()
```

```python
merged_df = pd.merge(scraped_df,concatenated_df,on='ASIN',how='left')
```

```python
merged_df
```

```python
merged_df.isnull().sum()
```

```python
merged_df['featureBullets_x'] = merged_df['featureBullets_x'].combine_first(merged_df['featureBullets_y'])
merged_df['productOverview_x'] = merged_df['productOverview_x'].combine_first(merged_df['productOverview_y'])
```

```python
merged_df.isnull().sum()
```

```python
scraped_df['productOverview'] = merged_df['productOverview_x']
scraped_df['featureBullets'] = merged_df['featureBullets_x']
```

```python
scraped_df[scraped_df['featureBullets'].isnull()]
```

```python
products_scraped_merged_df = pd.merge(products,scraped_df,on='ASIN',how='inner')
```

```python
gb_main_category = products_scraped_merged_df[products_scraped_merged_df['featureBullets'].isnull()].groupby('Main Category')
```

```python
gb_main_category.groups.keys()
```

```python
for name,gb_df in gb_main_category:
    print(name)
    print(gb_df.shape)
```

```python
gb_main_category.get_group('Beauty & Personal Care').sample(10)
```

```python
products_scraped_merged_df.isnull().sum()
```

```python
products_scraped_merged_df['Images'] = products_scraped_merged_df['Images'].apply(lambda item : ast.literal_eval(item))
```

```python
products_scraped_merged_df['productOverview'] = products_scraped_merged_df['productOverview'].apply(lambda item : ast.literal_eval(item) if pd.notna(item) else item)
products_scraped_merged_df['featureBullets'] = products_scraped_merged_df['featureBullets'].apply(lambda item : ast.literal_eval(item) if pd.notna(item) else item)
```

```python
print(products_scraped_merged_df.apply(lambda col: type(col.dropna().iloc[0]),axis=0))
```

```python
mostly_rescraped_books_df = pd.read_csv('MostlyBooksRescraped.csv')
mostly_rescraped_books_df.drop_duplicates(inplace=True)
```

```python
mostly_rescraped_books_df.drop(index=mostly_rescraped_books_df[mostly_rescraped_books_df['ASIN']=='B09X5WGSVG'].index,inplace=True)
```

```python
final_merge_df = pd.merge(products_scraped_merged_df,mostly_rescraped_books_df,on='ASIN',how='left')
```

```python
print(final_merge_df.apply(lambda col: type(col.dropna().iloc[0]),axis=0))
```
